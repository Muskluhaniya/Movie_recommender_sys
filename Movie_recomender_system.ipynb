```python
import numpy as np
import pandas as pd 
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
# displaying movies data 
movies.head(1)
```

```python
# displaying credits data 
credits.head(1)
```

```python
# merginf the data based on title
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
#we are keeping only these features
# genres
#id
#keywords
#title
#overview
#cast
#crew
```

```python
movies = movies[['genres','movie_id','keywords','title','overview','cast','crew']]
```

```python
movies.head(1)
```

```python
#check missing data
movies.isnull().sum()
```

```python
# 3 (missing values) is not much big number so we are just dropping it.
movies.dropna(inplace=True)
```

```python
#check duplicate values
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#convert any string into list 
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# I want to make tags so need this genres something like that ['Action','Adventure','Fantasy','Science Fiction']

def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
# need first 3 list of cast 
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
# need only director name from  crew
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break 
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
#overview is string, convert in list 
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
# we will remove all the space between words to avoid conflict in tags. 

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" " ,"") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" " ,"") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" " ,"") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" " ,"") for i in x])
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_df = movies[['movie_id','title','tags']]
new_df
```

```python
#need tags in strng not list
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head(1)
```

```python
new_df['tags'][0]
```

```python
# need in lowercase
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
#stemming 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# text vectorization 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
np.set_printoptions(threshold=np.inf)
cv.get_feature_names_out()
```

```python
# we will calculate the distance between movies but cosine distance not euclidean
#(in high dimension data, eucledian distance will fail and not a reliable measure)
# similarity is inv prop to distance

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)

```

```python
similarity[0]
```

```python
# we will find index of movie 
print(new_df[new_df['title'] == 'Avatar'].index)
print(new_df[new_df['title'] == 'Avatar'].index[0])
      
```

```python
#sorting of index
sorted(similarity[0])

#as we can see it is in asc order so we couldn't find the top 5 so we use reverse
sorted(similarity[0],reverse=True)

#but we are lossing index here, now we will use enumerate
list(enumerate(similarity[0]))

#now we will dor sorting again 
sorted(list(enumerate(similarity[0])),reverse=True,key= lambda x:x[1])[1:6] #we using key so it will sort list on the basis of distance not index
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key= lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title) #to get the movie name,i[0] is for index of that recommended movie
        
```

```python
recommend('Avatar')
```
